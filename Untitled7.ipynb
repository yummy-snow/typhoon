```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from sklearn.feature_selection import RFE
from sklearn.linear_model import LassoCV, LogisticRegression, LinearRegression
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import accuracy_score, f1_score, mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import TimeSeriesSplit
from xgboost import XGBClassifier, XGBRegressor
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
from haversine import haversine, Unit
```

```python
# 加载数据集
data_path = r"D:\环境数据分析\作业数据\typhoon_data.csv"  # 原始数据路径
df = pd.read_csv(data_path)

# 1. 处理缺失值
df['Maximum sustained wind speed'] = df['Maximum sustained wind speed'].fillna(df['Maximum sustained wind speed'].mean())
fields_to_fill = [
    'Direction of the longest radius of 50kt winds or greater',
    'The longeast radius of 50kt winds or greater',
    'The shortest radius of 50kt winds or greater',
    'Direction of the longest radius of 30kt winds or greater',
    'The longeast radius of 30kt winds or greater',
    'The shortest radius of 30kt winds or greater',
    'Indicator of landfall or passage'
]
df[fields_to_fill] = df[fields_to_fill].fillna(method='ffill')

# 合并时间字段为时间戳
df['datetime'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']])

# 经纬度标准化与异常值过滤
df = df[(df['Latitude of the center'] >= -90) & (df['Latitude of the center'] <= 90)]
df = df[(df['Longitude of the center'] >= -180) & (df['Longitude of the center'] <= 180)]

# 删除重复记录和无效数据
df = df.drop_duplicates()
df = df[(df['Central pressure'] > 0) & (df['Maximum sustained wind speed'] > 0)]

# 保存清理后的数据
cleaned_data_path = "cleaned_typhoon_data.csv"
df.to_csv(cleaned_data_path, index=False)

# -------------------- 探索性数据分析 (EDA) --------------------

# 统计描述
print("\n数值字段统计描述：")
numeric_stats = df.describe(percentiles=[0.25, 0.5, 0.75])
print(numeric_stats)

print("\n分类字段分布：")
category_stats = df['grade'].value_counts(normalize=True) * 100
print(category_stats)

# 台风等级分布可视化
plt.figure(figsize=(8, 6))
category_stats.plot(kind='bar', color='skyblue')
plt.title('Typhoon Grade Distribution')
plt.xlabel('Grade')
plt.ylabel('Percentage (%)')
plt.grid(axis='y')
plt.show()

# 台风路径图（folium）
print("\n绘制台风路径图...")
map_center = [df['Latitude of the center'].mean(), df['Longitude of the center'].mean()]
typhoon_map = folium.Map(location=map_center, zoom_start=5)

# 根据等级设置颜色
def get_color(grade):
    if grade == 'Tropical Depression':
        return 'blue'
    elif grade == 'Tropical Cyclone':
        return 'orange'
    else:
        return 'red'

for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row['Latitude of the center'], row['Longitude of the center']],
        radius=3,
        color=get_color(row['grade']),
        fill=True,
        fill_opacity=0.6
    ).add_to(typhoon_map)

typhoon_map.save("typhoon_path_map.html")
print("台风路径图已保存为 typhoon_path_map.html。")

# 时间趋势分析
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
yearly_counts = df.groupby('year')['grade'].count()

plt.figure(figsize=(10, 6))
yearly_counts.plot(kind='line', marker='o', color='green')
plt.title('Number of Typhoons per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.grid(True)
plt.show()

# 月份与台风强度的关系
plt.figure(figsize=(12, 6))
sns.boxplot(x='month', y='Maximum sustained wind speed', data=df, palette='coolwarm')
plt.title('Monthly Variation of Typhoon Intensity')
plt.xlabel('Month')
plt.ylabel('Maximum Sustained Wind Speed (kt)')
plt.grid(True)
plt.show()

# 特征关系分析
# 散点矩阵图
sns.pairplot(df[['Central pressure', 'Maximum sustained wind speed', 
                 'Latitude of the center', 'Longitude of the center']], diag_kind='kde')
plt.suptitle('Scatter Matrix of Features', y=1.02)
plt.show()

# 热力图显示相关性
corr_matrix = df[['Central pressure', 'Maximum sustained wind speed', 
                  'Latitude of the center', 'Longitude of the center']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Feature Correlations')
plt.show()

# -------------------- 特征工程 --------------------

# 1. 台风动力学特征
print("\n构造台风动力学特征...")

# 计算移动速度
df['prev_lat'] = df['Latitude of the center'].shift(1)
df['prev_lon'] = df['Longitude of the center'].shift(1)
df['prev_time'] = pd.to_datetime(df['datetime']).shift(1)

def calculate_speed(row):
    if pd.isnull(row['prev_lat']) or pd.isnull(row['prev_time']):
        return np.nan
    dist = haversine((row['prev_lat'], row['prev_lon']), (row['Latitude of the center'], row['Longitude of the center']), unit=Unit.KILOMETERS)
    time_diff = (pd.to_datetime(row['datetime']) - row['prev_time']).total_seconds() / 3600  # 时间差（小时）
    return dist / time_diff if time_diff > 0 else 0

df['speed_kmh'] = df.apply(calculate_speed, axis=1)

# 计算移动方向（方位角）
def calculate_direction(row):
    if pd.isnull(row['prev_lat']):
        return np.nan
    delta_lon = row['Longitude of the center'] - row['prev_lon']
    delta_lat = row['Latitude of the center'] - row['prev_lat']
    return np.degrees(np.arctan2(delta_lon, delta_lat))

df['direction_deg'] = df.apply(calculate_direction, axis=1)

# 2. 环境特征
print("\n构造环境特征...")

# 添加时间特征
df['quarter'] = pd.to_datetime(df['datetime']).dt.quarter
df['is_peak_month'] = df['month'].isin([7, 8, 9]).astype(int)  # 是否为台风高发月

# 3. 统计特征
print("\n构造统计特征...")

# 滑动窗口统计（过去6小时的平均气压变化率）
df['pressure_change_rate'] = df['Central pressure'].diff() / 6
df['rolling_pressure_rate'] = df['pressure_change_rate'].rolling(window=6, min_periods=1).mean()

# 检查新特征
print("\n新特征示例：")
print(df[['speed_kmh', 'direction_deg', 'quarter', 'is_peak_month', 'rolling_pressure_rate']].head())

# -------------------- 特征选择 --------------------

# 准备数据
X = df[['speed_kmh', 'direction_deg', 'quarter', 'is_peak_month', 'Central pressure', 'Maximum sustained wind speed', 
        'rolling_pressure_rate', 'Latitude of the center', 'Longitude of the center']].fillna(0)
y = df['Maximum sustained wind speed']

# 1. 过滤法：相关系数
print("\n过滤法：计算相关系数...")
correlations = X.corrwith(y).abs()
print(correlations)
selected_features_filter = correlations[correlations > 0.3].index.tolist()
print(f"过滤法选择的特征：{selected_features_filter}")

# 2. 包裹法：递归特征消除（RFE）
print("\n包裹法：递归特征消除（RFE）...")
rfe_model = RandomForestRegressor(random_state=42)
rfe = RFE(estimator=rfe_model, n_features_to_select=10, step=1)
rfe.fit(X, y)
selected_features_rfe = X.columns[rfe.support_].tolist()
print(f"包裹法选择的特征：{selected_features_rfe}")

# 3. 嵌入法：L1正则化（Lasso）
print("\n嵌入法：L1正则化（Lasso）...")
lasso = LassoCV(cv=5, random_state=42).fit(X, y)
selected_features_lasso = X.columns[(lasso.coef_ != 0)].tolist()
print(f"嵌入法选择的特征：{selected_features_lasso}")

# -------------------- 总结特征选择结果 --------------------
print("\n特征选择结果汇总：")
final_selected_features = set(selected_features_filter) | set(selected_features_rfe) | set(selected_features_lasso)
print(f"最终选择的特征：{final_selected_features}")

# -------------------- 可视化重要特征 --------------------

# 绘制相关系数热力图
plt.figure(figsize=(10, 8))
sns.heatmap(X.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Heatmap')
plt.show()

# 绘制Lasso系数
plt.figure(figsize=(10, 6))
plt.bar(X.columns, lasso.coef_)
plt.title('Lasso Coefficients')
plt.xticks(rotation=45)
plt.ylabel('Coefficient Value')
plt.grid(True)
plt.show()
```
